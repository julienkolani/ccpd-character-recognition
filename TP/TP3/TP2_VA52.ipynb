```python
import torch

# 1. Données d'entraînement
X = torch.tensor([[1.0, 2.0]])  # Entrée (1 exemple, 2 features)
Y = torch.tensor([[5.0]])       # Cible (La réponse que le modèle doit apprendre)

# 2. Poids et Biais (Les paramètres du modèle)
# requires_grad=True est essentiel pour la rétropropagation
w = torch.tensor([[0.5], [0.5]], requires_grad=True) # Poids (2x1)
b = torch.tensor([[0.1]], requires_grad=True)        # Biais (1x1)

# Hyperparamètres
learning_rate = 0.01  # Le taux d'apprentissage (pas d'ajustement)
```

```python
# Multiplier les entrées par les poids, puis ajouter le biais
Y_pred = torch.matmul(X, w) + b
print(f"Prédiction initiale (Y_pred): {Y_pred.item():.4f}")
# Pour notre exemple : (1*0.5 + 2*0.5) + 0.1 = 1.6
```

```python
loss = (Y_pred - Y).pow(2).mean()  # Calcul de la perte (MSE)
print(f"Perte initiale: {loss.item():.4f}")

loss.backward()  # Calcul des gradients
print(f"Gradient w: {w.grad.squeeze().numpy()}")
print(f"Gradient b: {b.grad.item():.4f}")
```

```python
with torch.no_grad():
    # Mise à jour des poids w
    w -= learning_rate * w.grad
    # Mise à jour du biais b
    b -= learning_rate * b.grad
    
    # Réinitialisation des gradients pour la prochaine itération
    w.grad.zero_()
    b.grad.zero_()
```

```python
print(f"\nNouveaux poids (w) après une étape: \n{w}")

# Nouvelle prédiction avec les poids ajustés
Y_pred_new = torch.matmul(X, w) + b
loss_new = (Y_pred_new - Y).pow(2).sum()

print(f"Nouvelle prédiction: {Y_pred_new.item():.4f}")
print(f"Nouvelle Perte: {loss_new.item():.4f}")
```

Calcul de ∂o/∂x :

1) o = mean(z) = (1/4) * Σ(zij)
2) z = (3 * y²)/x
3) y = x + 5

Donc :
∂o/∂x = (1/4) * Σ(∂z/∂x)
où ∂z/∂x = 3 * ((∂y²/∂x)/x - y²/x²)
et ∂y²/∂x = 2y * (∂y/∂x) = 2y * 1 = 2y

En remplaçant les valeurs :
y = x + 5 = 1 + 5 = 6
y² = 36
∂y²/∂x = 2 * 6 = 12

∂z/∂x = 3 * (12/1 - 36/1) = 3 * (-24) = -72

Donc ∂o/∂x = -72/4 = -18


```python

# Déclaration et initialisation d'un Tensor x de taille 2×2
x = torch.ones((2, 2), dtype=torch.float32, requires_grad=True)
print('x:', x)

# Déclaration de y = x + 5
y = x + 5
print('y:', y)

# Déclaration de z = (3 × y²)/x
z = (3 * y**2) / x
print('z:', z)

# Déclaration de o = mean(z)
o = z.mean()
print('o:', o)

# Calcul du gradient avec autograd
o.backward()
print('Gradient calculé par autograd (∂o/∂x):', x.grad)
```
